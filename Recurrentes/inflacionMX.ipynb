Se importan las librerías necesarias y se carga el archivo de datos de Banxico en un DataFrame para una primera inspección.

```python
import pandas as pd

df = pd.read_excel('Consulta_20251121-093940112.xlsx')
df.head(20)
```

Se identifica la fila que contiene los nombres de columnas y se extrae para usarla como encabezado del DataFrame.

```python
# Seleccionar la fila 8 como encabezado (índice 7)
new_header = df.iloc[8]
new_header
```

Se recortan filas de metadatos y se conserva la tabla útil a partir de la fila 17 del archivo original.

```python
# Crear un nuevo DataFrame desde la fila 17 en adelante
df = df[17:]
df
```

Se asignan los nombres de columnas usando el encabezado extraído y se verifica el resultado visualizando la cabecera.

```python
# Asignar el nuevo encabezado
df.columns = new_header
df.head()
```

Se reinicia el índice del DataFrame para que sea consecutivo y comience en cero.

```python
# Reiniciar el índice para que comience desde 0 en el nuevo DataFrame
df.reset_index(drop=True, inplace=True)
df.head()
```

Se convierte la columna de fecha desde texto a tipo datetime y se almacena en la nueva columna `Fecha`.

```python
# Convertir la columna 'fecha' a tipo datetime
df['Fecha'] = pd.to_datetime(df['Título'])
df.head()
```

Se genera un resumen informativo del DataFrame para revisar tipos de datos y valores faltantes.

```python
df.info()
```

Se filtra el rango temporal de interés (2014 a 2024) para crear el subconjunto de trabajo `_df`.

```python
# Filtrar por el año 2023
_df = df[(df['Fecha'] >= '01-01-2014') & (df['Fecha'] <= '12-31-2024')]
_df.head()
```

Se inspeccionan las últimas filas del subconjunto para confirmar el extremo más reciente del periodo seleccionado.

```python
_df.tail()
```

Se preparan las fechas y se grafica la variación anual del INPC para obtener una referencia visual de la serie temporal.

```python
import matplotlib.pyplot as plt

# dates = pd.to_datetime(_df["Fecha"]).apply(lambda x: x.strftime('%Y-%m-%d')).tolist()
dates = _df["Fecha"].tolist()

# Grafica el "Índice Nacional de Precios al consumidor, variación anual"
plt.figure(figsize=(16, 10))
plt.plot(dates,_df["Índice Nacional de Precios al consumidor, variación anual"], color='blue', label="Variación anual")
plt.title("Índice Nacional de Precios al Consumidor - variación anual")
plt.xlabel("Tiempo")
plt.ylabel("Variación anual")
plt.xticks(rotation=90)
plt.legend()
plt.show()
```

Se separan las características predictoras (`X`) y la variable objetivo (`y`), excluyendo columnas no útiles para el modelo.

```python
X = _df.drop(columns=["Índice Nacional de Precios al consumidor, variación anual", "Título", "Fecha"])
y = _df["Índice Nacional de Precios al consumidor, variación anual"].values
```

Se consulta la forma de `X` para verificar número de muestras y características.

```python
X.shape
```

Se consulta la forma de `y` para verificar el número de observaciones objetivo.

```python
y.shape
```

Se normalizan `X` y `y` al rango [0, 1] mediante `MinMaxScaler` para estabilizar el entrenamiento de la red.

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler(feature_range=(0, 1))
X_scaled = scaler.fit_transform(X)
y_scaled = scaler.fit_transform(y.reshape(-1, 1))
```

Se construyen ventanas temporales de tamaño fijo (`window_size`) para crear las secuencias de entrada de la RNN y sus etiquetas.

```python
import numpy as np

window_size = 12
X_rnn = []
y_rnn = []
# Construye secuencias temporales
for i in range(window_size, len(X_scaled)):
    X_rnn.append(X_scaled[i - window_size:i])
    y_rnn.append(y_scaled[i])

X_rnn, y_rnn = np.array(X_rnn), np.array(y_rnn)
```

Se divide el conjunto de datos en entrenamiento y prueba usando un 80% para entrenar y 20% para validar.

```python
split = int(len(X) * 0.8)
X_train, y_train = X_rnn[:split], y_rnn[:split]
X_test, y_test = X_rnn[split:], y_rnn[split:]
```

Se verifica la forma de `X_train` para confirmar la estructura (muestras, pasos de tiempo, características).

```python
X_train.shape
```

Se verifica la forma de `y_test` para confirmar el tamaño del conjunto de prueba.

```python
y_test.shape
```

Se visualiza el tamaño de la ventana temporal utilizada para generar las secuencias.

```python
window_size
```

Se consulta el número de características por paso temporal, necesario para definir la entrada del modelo.

```python
X_train.shape[2]
```

Se define una arquitectura de red recurrente simple con capas `SimpleRNN` y una capa densa de salida para regresión.

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, SimpleRNN, GRU

model = Sequential([
  SimpleRNN(units=60, return_sequences=True, input_shape=(window_size, X_train.shape[2])),
  SimpleRNN(units=30, return_sequences=True),
  SimpleRNN(units=15, return_sequences=False),
  Dense(units=1)
])
```

Se configura el optimizador Adam y se compila el modelo con función de pérdida MSE para un problema de regresión.

```python
from tensorflow.keras.optimizers import Adam

# Compilar el modelo
learning_rate = 0.001
adam_optimizer = Adam(learning_rate=learning_rate)
model.compile(optimizer=adam_optimizer, loss='mean_squared_error')
```

Se entrena el modelo con los datos de entrenamiento y se valida con el conjunto de prueba, definiendo épocas y tamaño de lote.

```python
# Entrenar el modelo
model.fit(X_train, y_train, batch_size=1, epochs=10, validation_data=(X_test, y_test))
```

Se generan las predicciones en el conjunto de prueba y se desescalan a la unidad original para facilitar la interpretación.

```python
# Predice en el conjunto de prueba y desescala los datos
predictions = model.predict(X_test)
predictions = scaler.inverse_transform(predictions)
y_test_original = scaler.inverse_transform(y_test)

# Muestra algunas predicciones y valores reales
print("Predicciones:", predictions.flatten()[:5])
print("Valores reales:", y_test_original.flatten()[:5])
```

Se calculan métricas de evaluación (MAE, MSE, RMSE y R²) para cuantificar el desempeño predictivo del modelo.

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import numpy as np

# Calcula el MAE
mae = mean_absolute_error(y_test, predictions)
print(f"Error Absoluto Medio (MAE): {mae}")

# Calcula el MSE
mse = mean_squared_error(y_test, predictions)
print(f"Error Cuadrático Medio (MSE): {mse}")

# Calcula el RMSE
rmse = np.sqrt(mse)
print(f"Raíz del Error Cuadrático Medio (RMSE): {rmse}")

# Calcula el Coeficiente de Determinación (R²)
r2 = r2_score(y_test, predictions)
print(f"Coeficiente de Determinación (R²): {r2}")
```

Se prepara un DataFrame de validación con fechas, valores reales y predicciones alineadas, y se grafica la comparación temporal.

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Gráfica de los valores reales y las predicciones del conjunto de prueba
valid = _df[split:]
valid = valid.reset_index(drop=True)
valid['Predictions'] = np.nan
valid.loc[window_size:, 'Predictions'] = predictions

dates_valid = pd.to_datetime(valid['Fecha']).apply(lambda x: x.strftime('%Y-%m-%d')).tolist()

plt.figure(figsize=(16, 8))
plt.title('Modelo RNN para Predicción de la inflación')
plt.xlabel('Fecha')
plt.ylabel('Inflacion')
plt.plot(dates_valid, valid[["Índice Nacional de Precios al consumidor, variación anual", 'Predictions']])
plt.legend(['Valor Real', 'Predicciones'], loc='lower right')
plt.xticks(rotation=90)
plt.show()
```
